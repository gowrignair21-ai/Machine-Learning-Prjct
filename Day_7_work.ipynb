```python
le=LabelEncoder() #label encoder is applied to target
df['body_mass_g']=le.fit_transform(df['body_mass_g'])
df.head()
```

```python
df_cleaned=df.dropna(subset=['body_mass_g'])

x=df_cleaned.drop(['body_mass_g'],axis=1)
y=df_cleaned['body_mass_g']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=12)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
df_cleaned
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
regressor=MLPRegressor(hidden_layer_sizes=(64,),activation='relu',solver='adam',random_state=42,max_iter=1000,verbose=10,learning_rate='adaptive',momentum=0.9)
regressor.fit(x_train,y_train)
```

```python
predictions=regressor.predict(x_test)
mse=mean_squared_error(y_test,predictions)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,predictions)
r2=r2_score(y_test,predictions)

print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'R-squared (r2) Score: {r2}')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input


model=Sequential()
model.add(Input(shape=(9,))) # input shape is 9 for df1 (diamonds dataset)
model.add(Dense(120,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(1,activation='linear')) # Changed to 1 output neuron for regression
```

```python
model.summary()
```

```python
#Compile
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
model.compile(optimizer='adam',loss='mse',metrics=['mae','mse','r2_score'])
```

```python
# Fit the model with the correctly prepared data
learned=model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=1, validation_data=(x_test,y_test))
loss, mae, mse, r2_score_val = model.evaluate(x_test, y_test)
print('Test mae:', mae)
print('Test mse:', mse)
print('Test r2_score:', r2_score_val)
```

```python
print('Test loss:',loss)
print('Test r2:',r2)
```

```python
plt.plot(learned.history['loss'],label='training loss')
plt.plot(learned.history['val_loss'],label='validation loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.show()
```

```python
df1[['x','y','z']]=df1[['x','y','z']].replace(0,np.nan)
df1.dropna(subset=['x','y','z'],inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ord_enc=OrdinalEncoder()
df1['cut']=ord_enc.fit_transform(df1[['cut']])
df1['color']=ord_enc.fit_transform(df1[['color']])
df1['clarity']=ord_enc.fit_transform(df1[['clarity']])
df1.head()
```

```python
columns=['cut','color','clarity','x','y','z']
for col in columns:
  df1[col]=np.log1p(df1[col])
df1.head()
```

```python
X = df1.drop('price', axis=1)
y = df1['price']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Standard scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input

model=Sequential()
model.add(Input(shape=(9,)))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['mae','mse','r2_score'])
learned=model.fit(x_train,y_train,epochs=100,batch_size=32,verbose=1,validation_data=(x_test,y_test))
```

```python
plt.plot(learned.history['loss'],label='training loss')
plt.plot(learned.history['val_loss'],label='validation loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.show()
```
